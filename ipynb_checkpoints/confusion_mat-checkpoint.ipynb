```python
 #Imports
import os
import glob
#if os.path.basename(os.getcwd()) == "notebooks": os.chdir("..")
import numpy as np
import matplotlib.pyplot as plt
import cv2 # this is a bit of an overkill to import for just one plot
import random
import colorsys
import scipy
import seaborn as sns


matrix1 = np.array([[56, 56], # load T F for each ds (do automatically read from validated output)
                    [1, 1]])

matrix2 = np.array([[8, 7],
                    [12, 13]])

labels = ['T', 'F']
columns = ['mot_en: Original', 'mot_en: avg']

# Create figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5), dpi=300)

sns.heatmap(matrix1, annot=True, fmt='d', cmap='Greens', xticklabels=columns, yticklabels=labels, ax=axs[0], cbar=False, linewidths=0.5)
axs[0].set_title('Twitch validation', fontsize=14)
axs[0].tick_params(axis='x', rotation=45)

sns.heatmap(matrix2, annot=True, fmt='d', cmap='Blues', xticklabels=columns, yticklabels=labels, ax=axs[1], cbar=False, linewidths=0.5)
axs[1].set_title('Active motion validation', fontsize=14)
axs[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
